```python
df = pd.read_csv('sharma-kohli.csv')
```

```python
df.head()
```

```python
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'], df['RG Sharma'])

plt.show()
```

# labeling titles

```python
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'], df['RG Sharma'])
plt.title("Comparing VK and RS IPL career score")
plt.xlabel('Year')
plt.ylabel("Runs Scored")

plt.show()
```

## chaning colors

```python
plt.plot(df['index'],df['V Kohli'], color='red',linestyle='dashed',lw='5')
plt.plot(df['index'], df['RG Sharma'], color='green',linestyle='dotted',lw='3')
plt.title("Comparing VK and RS IPL career score")
plt.xlabel('Year')
plt.ylabel("Runs Scored")

plt.show()
```

## markers

```python
plt.plot(df['index'],df['V Kohli'], color='red',linestyle='dashed',lw='2',marker='d',markersize='9')
plt.plot(df['index'], df['RG Sharma'], color='green',linestyle='dotted',lw='3')
plt.title("Comparing VK and RS IPL career score")
plt.xlabel('Year')
plt.ylabel("Runs Scored")

plt.show()
```

## labeling legends


```python
plt.plot(df['index'],df['V Kohli'], color='red',linestyle='dashed',lw='2',marker='d',markersize='9', label='Virat')
plt.plot(df['index'], df['RG Sharma'], color='green',linestyle='dotted',lw='3',label='Rohit')
plt.title("Comparing VK and RS IPL career score")
plt.xlabel('Year')
plt.ylabel("Runs Scored")
plt.legend()

plt.show()
```
