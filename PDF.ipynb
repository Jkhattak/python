```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
import random
```

```python
L= []

for i in range(1000):
    L.append(random.randint(1,6))
```

```python
l = (pd.Series(L).value_counts() / pd.Series(L).value_counts().sum()).sort_index()
```

```python
l.plot(kind=('bar'))
```
