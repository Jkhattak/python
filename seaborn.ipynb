```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
plt.style.use('ggplot')
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#This is axis level function
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
#replot 
sns.relplot(data=tips, x='total_bill',y='tip', kind='scatter', hue='sex',style='time',size='size')
```

```python
gap = px.data.gapminder()
```

```python
gap
```

```python
usa = gap[gap['country'] == 'United States']
```

```python
# axes level function

sns.lineplot(data=usa, x='year', y='lifeExp')
```

```python
# Figure level
sns.relplot(data=usa, x='year', y='lifeExp')
```

```python
threeNation = gap[gap['country'].isin(['Germany','China','United States'])]
```

```python
sns.relplot(data=threeNation, x='year',y='lifeExp', hue='country',kind='line',style='continent')
```

```python
sns.lineplot(data=threeNation, x='year',y='lifeExp', hue='country')
```

# facet plot

### <li> It will only work with figure style graphs

```python
sns.relplot(data=tips, x='total_bill',y='tip', col='sex')
```

## col wrap

```python
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='continent', col_wrap=1)
```

## Distribution plot

<li> Figure  level - > displot
<li> histplot - > kdeplot - > rugplot

```python
sns.histplot(data=tips,x='tip')
```

```python
sns.displot(data=tips,x='total_bill', kind='hist')
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
# titantic dataset

titanic = sns.load_dataset('titanic')
```

```python
sns.displot(data=titanic, x='age')
```

```python
sns.displot(data=titanic, x='age',element='step')
```

```python
sns.displot(data=titanic, x='age',element='step',hue='sex')
```

## kde plot 
<li> Kdeplot: A KDE plot smooths the observations with a Gaussian kernel rather than discrete bins, producing a continuous density 

```python
sns.kdeplot(data=tips, x='total_bill')

# the probability of 20 dollar total bill present in the data is about 0.05 percent
```
