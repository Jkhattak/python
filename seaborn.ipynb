## Categorical scatter plot

```python
# strip plot - > Takes one numerical and categorical columns. This is basically a scatter plot
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
```

```python
## figure levle function
## catplot - > is figure level function
sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

# Categorical Estimate plot - > Estimate some vlaue using these graphs 

```python
# Bar plot - When there are multiple observation in each category. It also uses bootstrapping to compute a
# confidence interval around the estimate, which is plotted using error bars

sns.barplot(data=tips, x='sex',y='total_bill')
```

```python
# hue parameter bar
sns.catplot(kind='bar', data=tips, x='sex',y='total_bill', hue='smoker')
```

```python
# point plot - > show the difference estimate
sns.catplot(kind='point', data=tips, x='sex',y='total_bill', hue='smoker')
```

```python
# count plot - > A special case for the bar plot is when you want to show th enumber of observations 
# in each category rahter than computing a statistic for second variable. This is similar to histogram over a categoricla, rather than quantitative variable

sns.countplot(data=tips,x='sex', hue='day')
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', col='smoker', kind='box', row='time')
```

# Regression plot
<li> scatter plot
<li> Regression plo
<li> In the simplest invocatino, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95 % confidence interval for that regression. 

```python
#shaded area represents error bar
# hue is not aviliable 
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
# residual plot
sns.residplot(data=tips, x='total_bill', y='tip')
```
